```python
import pandas as pd

messages = pd.read_csv('smsspamcollection.tsv', sep='\t')
messages = messages[['label', 'message']]
messages.head()
```

```python
# Data Cleaning and text preprocessing

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize

porter_stemmer = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
porter_stem_corpus =[]

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = word_tokenize(review)
    review = [porter_stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    porter_stem_corpus.append(review)

porter_stem_corpus
```

```python
porter_stem_corpus =[]

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.split()
    review = [porter_stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    porter_stem_corpus.append(review)

porter_stem_corpus
```

```python
lemmatized_corpus =[]

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.split()
    review = [wordnet_lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    lemmatized_corpus.append(review)

lemmatized_corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(max_features=100, binary=True)

porter_X = count_vectorizer.fit_transform(porter_stem_corpus).toarray()
lemmatized_X = count_vectorizer.fit_transform(lemmatized_corpus).toarray()
```
